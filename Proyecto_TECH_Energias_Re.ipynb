```python
# =====================================
# üì¶ Importaci√≥n de librer√≠as necesarias
# =====================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from io import StringIO
import requests

# =====================================
# üåê Fuente 1: Datos p√∫blicos desde datos.gov.co
# =====================================
print("üîΩ Cargando datos desde datos.gov.co...")
url_datosgov = "https://www.datos.gov.co/api/views/vy9n-w6hc/rows.csv?accessType=DOWNLOAD"

try:
    response = requests.get(url_datosgov)
    response.raise_for_status()  # Lanzar√° error si no es 200
    data_dgov = pd.read_csv(StringIO(response.text), encoding='utf-8')
    print("‚úÖ Datos cargados exitosamente desde datos.gov.co\n")
except Exception as e:
    print(f"‚ùå Error al cargar desde datos.gov.co: {e}\n")

# =====================================
# üåê Fuente 2: Archivo desde GitHub (CSV en tu repositorio p√∫blico)
# =====================================
print("üîΩ Cargando datos desde GitHub...")
url_github = "https://raw.githubusercontent.com/juan-manuel124/ProyectoAnalisisDatos-TECH/main/datos/proyectos_energia.csv"  # ajusta el path real si es necesario

try:
    data_git = pd.read_csv(url_github, encoding='utf-8')
    print("‚úÖ Datos cargados exitosamente desde GitHub\n")
except Exception as e:
    print(f"‚ùå Error al cargar desde GitHub: {e}\n")

# =====================================
# üîç Mostrar vista previa si ambos datasets cargaron
# =====================================
try:
    print("üìå Vista previa de datos de GitHub:")
    display(data_git.head())

    print("üìå Vista previa de datos de datos.gov.co:")
    display(data_dgov.head())
except:
    print("‚ö†Ô∏è Uno de los DataFrames no est√° disponible para mostrar.")
```

```python
# 2. Exploraci√≥n inicial
# =====================================
# üîç EXPLORACI√ìN INICIAL AVANZADA
# Documentacion para darle forma al codigo se encuentra en la libreriras de python
#https://docs.python.org/3/reference/lexical_analysis.html#f-strings
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.shape.html
# =====================================
def exploracion_mejorada(df, nombre_fuente="Dataset"):
    """
    Realiza una exploraci√≥n avanzada de un DataFrame con an√°lisis estad√≠sticos
    y visualizaciones clave para proyectos energ√©ticos.

    Args:
        df (pd.DataFrame): DataFrame a analizar
        nombre_fuente (str): Nombre descriptivo de la fuente de datos
    """
    # Configuraci√≥n inicial
    pd.set_option('display.float_format', '{:,.2f}'.format)
    print(f"\n{'='*80}\nüîç EXPLORACI√ìN AVANZADA: {nombre_fuente.upper()}\n{'='*80}")#otiene la fuente (tipo letra )

    # 1. Metadatos b√°sicos
    print(f"\nüìå METADATOS [Filas: {df.shape[0]} | Columnas: {df.shape[1]}]")
    print(f"üìä Memoria usada: {df.memory_usage(deep=True).sum()/1024**2:.2f} MB") #muestra el uso de memoria para detectar posibles desbordamientos

    # 2. Tipos de datos y completitud
    # Este bloque de c√≥digo analiza los tipos de datos de cada columna,
    # cuenta los valores no nulos, calcula el porcentaje de completitud
    # y determina la cantidad de valores √∫nicos por columna.

    # Se crea un nuevo DataFrame llamado 'tipos_df'.
    # Este DataFrame contendr√° el resumen del an√°lisis de los datos.
    tipos_df = pd.DataFrame({
        # 'Tipo': Se crea una columna en 'tipos_df' llamada 'Tipo'.
        # df.dtypes: Esto obtiene el tipo de dato de cada columna en el DataFrame original 'df'.
        'Tipo': df.dtypes,

        # 'No Nulos': Se crea una columna en 'tipos_df' llamada 'No Nulos'.
        # df.count(): Esto cuenta el n√∫mero de valores no nulos (no NaN) en cada columna de 'df'.
        'No Nulos': df.count(),

        # '% Completitud': Se crea una columna para mostrar el porcentaje de valores no nulos.
        # df.count() / len(df): Calcula la proporci√≥n de valores no nulos dividiendo
        #                      el conteo de no nulos por el n√∫mero total de filas (len(df)).
        # * 100: Multiplica la proporci√≥n por 100 para obtener un porcentaje.
        # .round(2): Redondea el porcentaje a 2 decimales.
        '% Completitud': (df.count() / len(df) * 100).round(2),

        # 'Valores √önicos': Se crea una columna para mostrar la cantidad de valores √∫nicos.
        # df.nunique(): Esto cuenta el n√∫mero de valores √∫nicos en cada columna de 'df'.
        'Valores √önicos': df.nunique()
    })

    # Despu√©s de ejecutar este c√≥digo, 'tipos_df' ser√° un DataFrame
    # donde cada fila representa una columna del DataFrame original 'df',
    # y las columnas de 'tipos_df' ('Tipo', 'No Nulos', '% Completitud', 'Valores √önicos')
    # contendr√°n la informaci√≥n resumida para cada columna original.

    print("\nüßæ RESUMEN DE VARIABLES:")
    display(tipos_df)

    # 3. An√°lisis de variables num√©ricas
    # Selecciona solo las columnas num√©ricas del DataFrame
    num_cols = df.select_dtypes(include=np.number).columns

    # Si hay al menos una columna num√©rica
    if len(num_cols) > 0:
        print("\nüî¢ ESTAD√çSTICAS NUM√âRICAS:")

        # Muestra estad√≠sticas descriptivas detalladas (incluye percentiles extremos)
        display(df[num_cols].describe(percentiles=[.01, .25, .5, .75, .99]).transpose())

        # --- Detecci√≥n de valores at√≠picos (outliers) usando el criterio IQR ---

        # Calcula el primer cuartil (Q1) para cada variable num√©rica
        Q1 = df[num_cols].quantile(0.25)

        # Calcula el tercer cuartil (Q3) para cada variable num√©rica
        Q3 = df[num_cols].quantile(0.75)

        # Calcula el rango intercuart√≠lico (IQR)
        IQR = Q3 - Q1

        # Identifica cu√°ntos valores est√°n fuera del rango normal:
        #   - Menores a Q1 - 1.5 * IQR (posibles valores muy bajos)
        #   - Mayores a Q3 + 1.5 * IQR (posibles valores muy altos)
        outliers = ((df[num_cols] < (Q1 - 1.5 * IQR)) | (df[num_cols] > (Q3 + 1.5 * IQR))).sum()

        print("\n‚ö†Ô∏è POTENCIALES OUTLIERS (criterio IQR):")

        # Muestra el conteo de outliers por cada columna num√©rica
        display(outliers.to_frame('Conteo'))


    # 4. An√°lisis de variables categ√≥ricas
    cat_cols = df.select_dtypes(exclude=np.number).columns
    if len(cat_cols) > 0:
        print("\nüî§ DISTRIBUCI√ìN CATEG√ìRICA:")
        for col in cat_cols:
            if df[col].nunique() < 20:
                print(f"\nüìä {col}:")
                display(df[col].value_counts(normalize=True).head(10))

    # 5. Visualizaciones clave para energ√≠a
    if 'Energ√≠a [kWh/d√≠a]' in df.columns:
        plt.figure(figsize=(12, 6))
        sns.histplot(np.log10(df['Energ√≠a [kWh/d√≠a]']+1), bins=30, kde=True)
        plt.title('Distribuci√≥n Logar√≠tmica de Energ√≠a Diaria')
        plt.xlabel('log10(Energ√≠a [kWh/d√≠a])')
        plt.show()

    if 'Tipo' in df.columns:
        plt.figure(figsize=(8, 5))
        df['Tipo'].value_counts().plot(kind='pie', autopct='%1.1f%%')
        plt.title('Distribuci√≥n de Tipos de Proyectos')
        plt.ylabel('')
        plt.show()

    # 6. Correlaciones num√©ricas
    if len(num_cols) > 1:
        plt.figure(figsize=(10, 8))
        sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm', center=0)
        plt.title('Matriz de Correlaci√≥n')
        plt.show()

# =====================================
# üöÄ EJECUCI√ìN PARA DATOS CARGADOS
# =====================================
if 'data_dgov' in locals():
    print("\n" + "="*60)
    print("üìä AN√ÅLISIS DATOS.GOV.CO")
    print("="*60)
    exploracion_mejorada(data_dgov, "Datos P√∫blicos Energ√≠a")

if 'data_git' in locals():
    print("\n" + "="*60)
    print("üìä AN√ÅLISIS DATOS GITHUB")
    print("="*60)
    exploracion_mejorada(data_git, "Proyectos Energ√≠a GitHub")

```

```python
# 4. An√°lisis por departamento
print("\n" + "="*80)
print("üîç AN√ÅLISIS POR DEPARTAMENTO ‚Äî Enfoque PERS (Producci√≥n, Empleo, Recursos, Servicios)")
print("="*80)

# Agrupaci√≥n agregada por departamento
resumen_por_depto = (
    df_clean
    .groupby('Departamento')
    .agg({
        'Energ√≠a [kWh/d√≠a]': 'sum',
        'Usuarios': 'sum',
        'Capacidad': 'sum',
        'Inversi√≥n estimada [COP]': 'sum',
        'Empleos estimados': 'sum'
    })
    .rename(columns={
        'Energ√≠a [kWh/d√≠a]': 'üîã Consumo Total (kWh/d√≠a)',
        'Usuarios': 'üë• Total Usuarios',
        'Capacidad': '‚ö° Capacidad Instalada (kW)',
        'Inversi√≥n estimada [COP]': 'üí∞ Inversi√≥n Total (COP)',
        'Empleos estimados': 'üë∑ Empleos Estimados'
    })
    .sort_values(by='üîã Consumo Total (kWh/d√≠a)', ascending=False)
)

# Visualizaci√≥n tabular en consola
print("\nüìä Resumen energ√©tico y social por departamento:")
display(resumen_por_depto.style.format({
    'üîã Consumo Total (kWh/d√≠a)': '{:,.0f}',
    'üë• Total Usuarios': '{:,.0f}',
    '‚ö° Capacidad Instalada (kW)': '{:,.2f}',
    'üí∞ Inversi√≥n Total (COP)': '${:,.0f}',
    'üë∑ Empleos Estimados': '{:,.0f}'
}).set_table_styles([{'selector': 'th', 'props': [('text-align', 'center')]}]))

# Exportar si se desea (opcional)
# resumen_por_depto.to_csv("resumen_departamento_pers.csv")

print("\n‚úÖ An√°lisis por departamento completado.")
print("="*80)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

print("\n Iniciando an√°lisis temporal visual con Seaborn...\n")

# Estilo visual
sns.set_theme(style="whitegrid", palette="deep")

# =============================
# Preparaci√≥n de columnas temporales
# =============================
df_clean['Periodo'] = df_clean['Fecha estimada FPO'].dt.to_period('M')
df_clean['A√±o'] = df_clean['Fecha estimada FPO'].dt.year

# =============================
# Agregaciones necesarias
# =============================
df_anual = df_clean.groupby('A√±o', observed=False).agg({
    'Energ√≠a [kWh/d√≠a]': 'sum',
    'Proyecto': 'count',
    'Inversi√≥n estimada [COP]': 'sum',
    'Empleos estimados': 'sum'
}).reset_index().rename(columns={
    'Proyecto': 'Cantidad de Proyectos',
    'Energ√≠a [kWh/d√≠a]': 'Energ√≠a Total',
    'Inversi√≥n estimada [COP]': 'Inversi√≥n Total',
    'Empleos estimados': 'Empleos Estimados'
})

df_mensual = df_clean.groupby('Periodo', observed=False)['Energ√≠a [kWh/d√≠a]'].sum().reset_index()
df_mensual['Periodo'] = df_mensual['Periodo'].astype(str)

import matplotlib.pyplot as plt
import seaborn as sns

# Estilo profesional
sns.set_theme(style='whitegrid', palette='deep')

# =============================
# Subgr√°ficos anuales organizados din√°micamente
# =============================
# Definimos los indicadores y estilos
indicadores = {
    'Energ√≠a Total': {'palette': 'flare', 'tipo': 'bar', 'ylabel': 'kWh/d√≠a'},
    'Cantidad de Proyectos': {'palette': 'crest', 'tipo': 'bar', 'ylabel': 'Cantidad'},
    'Inversi√≥n Total': {'color': 'green', 'tipo': 'line', 'ylabel': 'COP'},
    'Empleos Estimados': {'palette': 'coolwarm', 'tipo': 'bar', 'ylabel': 'Empleos'}
}

fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('Indicadores Anuales del Proyecto', fontsize=18, y=1.02)

for ax, (col, props) in zip(axes.flat, indicadores.items()):
    if props['tipo'] == 'bar':
        sns.barplot(data=df_anual, x='A√±o', y=col, palette=props['palette'], ax=ax)
    elif props['tipo'] == 'line':
        sns.lineplot(data=df_anual, x='A√±o', y=col, marker='o', color=props['color'], ax=ax)

    # L√≠neas estad√≠sticas de referencia
    mean_val = df_anual[col].mean()
    std_val = df_anual[col].std()
    ax.axhline(mean_val, color='blue', linestyle='dashed', label='Media')
    ax.axhline(mean_val + std_val, color='purple', linestyle='--', label='+1œÉ')
    ax.axhline(mean_val - std_val, color='red', linestyle='--', label='‚àí1œÉ')

    ax.set_title(f'{col} por A√±o')
    ax.set_ylabel(props['ylabel'])
    ax.set_xlabel('')
    ax.legend(fontsize=8, loc='upper right')
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
